# Core Components of an Agentic AI (with real-time examples)

1. Brain / Reasoning Engine (LLM)
    - Role: Plan, decompose goals, make decisions.
    - Real-time example: LLM analyzes a hiring brief, generates a multi-step sourcing plan and ranks candidate-fit reasons.

2. Orchestrator / Workflow Manager
    - Role: Sequence tasks, monitor progress, retry/fallback.
    - Real-time example: Orchestrator triggers resume-parsing → shortlist → schedule interview steps and retries calendar API on failure.

3. Tools / Integrations
    - Role: External capabilities (APIs, parsers, UIs).
    - Real-time example: Resume parser extracts skills, Calendar API books slots, Email API sends interview invites, Job board APIs post JDs.

4. Memory (Short-term & Long-term)
    - Role: Store context, conversation state, preferences, past outcomes.
    - Real-time example: Short-term memory holds current candidate shortlist; long-term memory stores hiring manager preferences (remote-first, seniority) to bias future searches.

5. Supervisor / Safety & Human-in-the-Loop
    - Role: Policies, approvals, overrides, audit logs.
    - Real-time example: Supervisor blocks offers above budget automatically and routes flagged hires to a human approver; audit trail records all actions.

6. Sensors & Observability
    - Role: Monitor system state, tool responses, external signals.
    - Real-time example: Dashboard shows applicant flow, API latencies, and alerts if applicant rate drops below threshold.

7. Feedback & Learning Loop
    - Role: Use outcomes to improve planning, ranking, and policies.
    - Real-time example: After interviews, hire/no-hire labels are fed back to update candidate-scoring heuristics.

Quick composite scenario:
- User posts a job → Brain creates plan → Orchestrator runs tools (post JD, parse resumes) → Memory retains context → Supervisor enforces guardrails (no offers without approval) → Observability alerts on issues → Feedback updates scoring after hires.

# What is Langgraph
* Langgraph is a orchestration frameworks for buliding inteligence, statefull, and Multisteo LLM Workflow

* It enable advance Feature like parallelism ,loop, branching,memory and Resumablity-making it ideal for agentic and production -grade Ai application
 
* It models your logic as a graph of nodes (tasks) and edges(routing) instead of a linear chian.

* # Core concepts in langgraph
1. LLM Workflows
2. Graphs, Nodes, and Edges 
3. State
4. Reducers
5. LangGraph Execution Model 


1. LLM Workflows (5:43):

* * LLM workflow are a steo by step process using which we can build complex LLM application.

* * Each step in a workflow perform a distinct task -- such as prompting, reasoning, Tool calling, Memory access, or Decision making.

* *  Workflow can be linear, parallel, Branching, or loop allowing for complex behaviors like retries, multi-agent communication ,or Tool-augmented reasoning.

* * Common workflow


* Prompt Chaining :
* *  Sequential LLM calls for complex tasks.

* Routing :
* *  Directing queries to specific LLMs based on task nature.

* Parallelization : 
* * Executing sub-tasks simultaneously.

* Orchestrator-Worker :
* * Dynamic task assignment via an orchestrator.

* Evaluator-Optimizer :
* * Iterative refinement of outputs.

2. Graphs, Nodes, and Edges : 
* * Explains that LangGraph represents workflows as graphs where Nodes are Python functions (tasks) and Edges determine the flow of execution (sequence, branching, loops).

3. State : 
* * In langgraph ,State is the shared memory that flows though your workflow-- it hold all the data being passed between nodes as your graph runs. 


4. Reducers :
* * Reducer in langgraph define how update from nodes are applied to the shared state.
* * Each key in the state can have its own reducer, Which determines whether new data replace, merges,or adds to the existing value.


5. LangGraph Execution Model : 
* * Details the behind-the-scenes mechanics inspired by Google Pregel. It covers the transition from graph definition to compilation and the use of message passing within supersteps to execute workflows efficiently.

```python
from dataclasses import dataclass
from typing import List, Dict, Any
from datetime import datetime
from enum import Enum

# Define component classes
class TaskStatus(Enum):
    PENDING = "pending"
    IN_PROGRESS = "in_progress"
    COMPLETED = "completed"
    FAILED = "failed"

@dataclass
class Task:
    id: str
    name: str
    status: TaskStatus = TaskStatus.PENDING
    result: Any = None

class Brain:
    """Reasoning Engine - Plans and makes decisions"""
    def plan(self, goal: str) -> List[str]:
        return [f"Step 1: Analyze {goal}", f"Step 2: Generate candidates", f"Step 3: Evaluate results"]
    
    def reason(self, context: Dict) -> str:
        return f"Decision based on context: {context}"

class Orchestrator:
    """Workflow Manager - Sequences and monitors tasks"""
    def __init__(self):
        self.task_queue = []
    
    def add_task(self, task: Task):
        self.task_queue.append(task)
    
    def execute_workflow(self) -> List[Task]:
        for task in self.task_queue:
            task.status = TaskStatus.IN_PROGRESS
            task.result = f"Executed {task.name}"
            task.status = TaskStatus.COMPLETED
        return self.task_queue

class Tools:
    """External Integrations - APIs and parsers"""
    def parse_resume(self, resume_data: str) -> Dict:
        return {"skills": ["Python", "AWS"], "experience": "2 years"}
    
    def post_job_description(self, jd: str, platform: str) -> bool:
        return True
    
    def schedule_interview(self, candidate: str, time: str) -> bool:
        return True

class Memory:
    """Short-term & Long-term context storage"""
    def __init__(self):
        self.short_term = {}
        self.long_term = {}
    
    def store_short_term(self, key: str, value: Any):
        self.short_term[key] = value
    
    def store_long_term(self, key: str, value: Any):
        self.long_term[key] = value
    
    def retrieve(self, key: str) -> Any:
        return self.short_term.get(key) or self.long_term.get(key)

class Supervisor:
    """Safety, Policies & Human-in-the-Loop"""
    def __init__(self):
        self.guardrails = {"max_salary": 150000, "require_approval": True}
        self.audit_log = []
    
    def check_guardrail(self, action: str, params: Dict) -> bool:
        self.audit_log.append({"action": action, "timestamp": datetime.now()})
        if action == "send_offer" and params.get("salary", 0) > self.guardrails["max_salary"]:
            return False
        return True

# Initialize Agentic AI System
brain = Brain()
orchestrator = Orchestrator()
tools = Tools()
memory = Memory()
supervisor = Supervisor()

# Real-time example: Hiring workflow
print("🤖 Agentic AI Hiring System\n")
goal = "Hire Senior Python Developer"
print(f"Goal: {goal}")
print(f"Plan: {brain.plan(goal)}\n")

# Store context in memory
memory.store_short_term("current_candidates", ["Alice", "Bob", "Charlie"])
memory.store_long_term("hiring_preferences", {"remote_first": True, "seniority": "Senior"})

# Execute workflow
task1 = Task(id="1", name="Post Job Description")
task2 = Task(id="2", name="Parse Resumes")
orchestrator.add_task(task1)
orchestrator.add_task(task2)

completed_tasks = orchestrator.execute_workflow()
print(f"Completed Tasks: {len(completed_tasks)}")

# Apply guardrails
offer_params = {"salary": 120000, "candidate": "Alice"}
if supervisor.check_guardrail("send_offer", offer_params):
    print(f"✅ Offer approved: {offer_params}\n")

print(f"Audit Log: {len(supervisor.audit_log)} actions recorded")
```
