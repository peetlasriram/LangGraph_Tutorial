```python
from langgraph.graph import StateGraph, START, END
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from typing import TypedDict
```

```python
model = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct", temperature=0.7)
```

```python
class BlogState(TypedDict):
    title:str
    outline:str
    content:str
```

```python
def create_outline(state:BlogState)-> BlogState:
    title=state["title"]
    prompt=f"Generate detailed outline for a blong on the topic   {title}"
    result= model.invoke(prompt).content
    state["outline"]= result
    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    title = state['title']
    outline = state['outline']

    prompt = f'Write a detailed blog on the title - {title} using the follwing outline \n {outline}'

    content = model.invoke(prompt).content

    state['content'] = content

    return state
```

```python
# Define Graph
graph= StateGraph(BlogState)

# Define Nodes
graph.add_node("create_outline",create_outline)
graph.add_node("create_blog",create_blog)

# Define Edges
graph.add_edge(START ,"create_outline")
graph.add_edge("create_outline","create_blog")
graph.add_edge("create_blog",END)
# Compile model

work_flow= graph.compile()
```

```python
intial_state = {'title': 'Rise of AI in India'}

final_state = work_flow.invoke(intial_state)

print(final_state)

```
